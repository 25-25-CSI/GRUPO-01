```python
import os
import lorem

# Definimos la ruta donde se guardarán los archivos
save_path = "/content/drive/MyDrive/CriptografiaySeguridad/CriptoDeber2/Archivos"
os.makedirs(save_path, exist_ok=True)

# Definimos las cantidades de palabras
word_counts = [10, 100, 1000, 10000, 100000, 1000000, 10000000]

for count in word_counts:
    words = []
    while len(words) < count:
        words.extend(lorem.text().split())

    # Cortamos exactamente el número de palabras necesarias
    words = words[:count]

    # Convertimos la lista de palabras a una sola cadena
    content = ' '.join(words)

    # Definimos nombre y ruta
    filename = f"{count}_palabras.txt"
    filepath = os.path.join(save_path, filename)

    # Guardamos el archivo
    with open(filepath, "w", encoding="utf-8") as f:
        f.write(content)

    print(f"Archivo {filename} generado con {count} palabras en {save_path}")
```
