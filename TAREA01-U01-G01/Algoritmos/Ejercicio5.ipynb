# Método de sustitución Polialfabético de Vigenère.

Algoritmo que realice el cifrado de una cadena de caracteres
mediante un método de sustitución Polialfabético de Vigenère.
La cadena se ingresa al iniciar el algoritmo. El algoritmo debe
 mostrar la cadena de caracteres ingresada, la clave de cifrado
 y la cadena de caracteres cifrada.

 Bloque 1 – Función para generar la clave

Permite cifrar/desencriptar manteniendo la alineación con el mensaje original, respetando los espacios.

```python
def generate_key(message, key_word):
    key = []
    key_word_count = 0
    for x in range(0, len(message)):
        if message[x] == " ":
            key.append(" ")
        else:
            i = key_word_count % len(key_word)
            key.append(key_word[i])
            key_word_count += 1
    key = ''.join(str(i) for i in key)
    return key
```

Bloque 2 – Función para encriptar con Vigenère

Convierte el mensaje original en un texto cifrado que no es legible a simple vista.

```python
def encrypt_vigenere(message, key_word):
    key = generate_key(message, key_word)
    encrypted_text = ""

    for i in range(0, len(message)):
        if message[i] == " ":
            encrypted_text += " "
        else:
            encryption_value = (alphabets.find(message[i]) + alphabets.find(key[i])) % 26
            encrypted_text += alphabets[encryption_value]

    return encrypted_text
```

Bloque 3 – Función para desencriptar con Vigenère

Recupera el mensaje original a partir del texto cifrado y la misma clave usada para cifrar

```python
def decrypt_vigenere(message, key_word):
    key = generate_key(message, key_word)
    decrypted_text = ""

    for i in range(0, len(message)):
        if message[i] == " ":
            decrypted_text += " "
        else:
            decryption_value = (alphabets.find(message[i]) - alphabets.find(key[i])) % 26
            decrypted_text += alphabets[decryption_value]

    return decrypted_text
```

Bloque 4 – Definición del alfabeto

```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

Bloque 5 – Función principal

Permite convertir letras en números y viceversa durante el cifrado y descifrado

```python
def main():
    message = str(input("Escribe tu mensaje: ")).upper()
    key_word = str(input("Palabra Clave: ")).upper()

    print("El mensaje es: " + str(message))

    key = generate_key(message, key_word)
    print("Clave_Vigenère: " + str(key))

    encrypted_text = encrypt_vigenere(message, key_word)
    print("Criptado_Vigenère: " + str(encrypted_text))

    decrypted_text = decrypt_vigenere(encrypted_text, key_word)
    print("Desencriptado_Vigenère: " + str(decrypted_text))
```

Bloque 6 – Llamada al main


```python
if __name__ == '__main__':
    main()
```
