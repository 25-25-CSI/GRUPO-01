# Método de sustitución Monoalfabético de desplazamiento n caracteres a la derecha

4. Algoritmo que realice el cifrado de una cadena de caracteres mediante un método de sustitución Monoalfabético de desplazamiento n caracteres a la derecha. Tanto la palabra como el valor de n se ingresan al iniciar el algoritmo. El algoritmo debe mostrar el alfabeto original, el alfabeto cifrado, la cadena de caracteres ingresada y su resultado.

```python
alfabeto = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
def crear_alfabetos(desplazamiento):
    # Devuelve una tupla (alfabeto_original, alfabeto_cifrado) desplazado n posiciones.
    desplazamiento %= len(alfabeto)  # Asegura que n esté en [0, 25]
    alfabeto_cifrado = alfabeto[-desplazamiento:] + alfabeto[:-desplazamiento]
    return alfabeto, alfabeto_cifrado
```

```python
def cifrar_texto(texto, alfabeto, alfabeto_cifrado):
    # Cifra el texto usando los alfabetos dado el mapeo de sustitución.
    resultado = []
    for ch in texto:
        ch_upper = ch.upper()
        if ch_upper in alfabeto:
            i = alfabeto.index(ch_upper)
            # Conserva mayúsculas/minúsculas
            nuevo = alfabeto_cifrado[i]
            resultado.append(nuevo if ch.isupper() else nuevo.lower())
        else:
            resultado.append(ch)
    return ''.join(resultado)
```

```python
def main():
    # Entrada
    texto = input("Ingresa la cadena a cifrar: ")
    try:
        n = int(input("Ingresa el valor de desplazamiento n (entero): "))
    except ValueError:
        print("Desplazamiento inválido. Usa un número entero.")
        return

    # Generar alfabetos
    alfabeto_orig, alfabeto_cif = crear_alfabetos(n)

    # Cifrar
    texto_cifrado = cifrar_texto(texto, alfabeto_orig, alfabeto_cif)

    # Salida
    print(f"\nAlfabeto original: {alfabeto_orig}")
    print(f"Alfabeto cifrado : {alfabeto_cif}")
    print(f"Texto ingresado  : {texto}")
    print(f"Texto cifrado    : {texto_cifrado}")

if __name__ == "__main__":
    main()
```
